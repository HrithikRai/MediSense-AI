```python
import pandas as pd
```

```python
drugs = pd.read_csv("data/raw/treatment_data.csv")
```

```python
drugs
```

```python
medicine_data = pd.read_csv("data/raw/indian_medicine_data.csv")
```

```python
medicine_data
```

```python
medicine_data['short_composition1'] = medicine_data['short_composition1'].str.split(' ').str[0]
```

```python
medicine_data
```

```python
# calculate price/tablet
medicine_data['pack_size_label'] =  medicine_data['pack_size_label'].str.extract('(\d+)').astype(int)
```

```python
medicine_data.head(2)
```

```python
medicine_data['price_per_tablet'] = medicine_data['price(₹)'] / medicine_data['pack_size_label']
```

```python
medicine_data.head(2)
```

```python
medicine_data.drop(medicine_data[medicine_data["Is_discontinued"] == True].index, inplace=True)
```

```python
drugs.head(2)
```

```python
medicine_data.head(2)
```

```python
medicine_data.rename(columns={'short_composition1':'drug_name'},inplace=True)
```

```python
# merge dataframes
doctor = pd.merge(drugs, medicine_data[['name','drug_name','price_per_tablet']], on='drug_name')
```

```python
doctor.head(20)
```

```python
doctor = doctor.drop_duplicates(['drug_name','type','disease','symptom'], keep='first')
```

```python
doctor.reset_index(inplace=True)
```

```python
doctor.head()
```

```python
test = doctor.groupby('drug_name', as_index=False).agg({'symptom': ', '.join})
```

```python
doctor = doctor.drop_duplicates(['drug_name','type','disease'], keep='first')
```

```python
doctor.drop(columns=['symptom'],axis=1,inplace=True)
```

```python
doctor
```

```python
test
```

```python
final = pd.merge(doctor, test, on='drug_name')
```
